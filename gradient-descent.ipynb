```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sb
tips = sb.load_dataset('tips')
tips.head(5)
X = tips['total_bill']
Y = tips['tip']
plt.scatter(X,Y)
plt.show()
```

```python
m = 0
c = 0
Y_pred = m*X + c
plt.plot([min(X),max(X)],[min(Y_pred),max(Y_pred)],color = 'red')
plt.scatter(X,Y)
```

```python
L = 0.0001
iter_ = 1000

n = float(len(X))
print(n)
#For loop - Performing Gradient Descent 
for i in range(iter_):
    Y_pred = (m*X)+c #The current prediction value 
    D_m = (-2/n)*sum(X * (Y - Y_pred))
    D_c = (-2/n)*sum(Y - Y_pred)
    m = m - (L*D_m) 
    c = c - (L*D_c)
    print("The Slope and intercept for iteration:",i+1)
    print("Slope: ",m)
    print("Intercept: ",c)
    
print("After " +  str(iter_) + " iterations value of slope: %0.2f " %(m))
print("After " +  str(iter_) + " iterations value of intercept: %0.2f " %(c))
```

```python
Y_pred = m*X + c

plt.scatter(X,Y)
plt.plot([min(X),max(X)],[min(Y_pred),max(Y_pred)],color = 'red')
plt.show()
```
